```python
nike_data_ext = nike_data[['tweet_created_at', 'user_name','user_location', 'user_description','user_following',
                           'user_friends_count','user_followers_count','user_following', 'tweet_full_text',
                           'user_default_profile']]
```

```python
nike_data_ext.head(10)
```

```python
fig, ax = plt.subplots(figsize=(12,10))
nike_data_ext.user_location.value_counts().sort_values(ascending=True).tail(40).plot(kind="barh")
plt.ylabel("Locations of users")
plt.xlabel("Counts")
plt.title("User Counts by location")
plt.show()
```
