```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

According to a 2022 report by Statista, the top 5 sportswear brands by market share are:

1. Nike (49.0%)
2. Adidas (23.0%)
3. Puma (5.0%)
4. Under Armour (4.0%)
5. Lululemon Athletica (3.0%)
6. New Balance (2.0%)
7. Asics (2.0%)
8. Reebok (1.0%)
9. Columbia Sportswear (1.0%)
10. Fila (1.0%)

These brands account for over 90% of the global sportswear market.

It is important to note that the market share of these brands can vary depending on the specific product category and geographic region. For example, Nike is the dominant player in the global sportswear market as a whole, but Adidas has a larger market share in the European soccer market.

Reference: https://www.statista.com/statistics/900271/leading-sportswear-and-performance-wear-companies-by-sales-worldwide/

```python
data = pd.DataFrame(
    {
        "Brand": ["Nike", "Adidas", "Puma", "Under Armour", "Lululemon Athletica", "New Balance", "Asics", "Reebok",
                 "Columbia Sportswear", "Fila"],
        "Share": [49, 23, 5, 4, 3, 2, 2, 1, 1, 1]
    }
)
data
```

```python
data = data.head()
fig, ax = plt.subplots(figsize=(6, 6))
plt.pie(data["Share"], labels=data["Brand"], autopct="%1.1f%%")
plt.title("Sportwears Market share by Percentage", color="brown", fontsize=12)
plt.annotate(
    "Statistica (2023)",
    xy=(0.5,-1.2),
    xytext=(7, 0.8),
    textcoords="offset points",
    fontsize=10,
    color="brown"
)

plt.show()
```
